```python
df_east = pd.read_csv('./data/news_eastmoney.csv')
```

```python
###查看标题情感标签（sentiment）
###有三种标签：”利好“，”利空“，”中性“

df_east['sentiment'] = 0
for i, row in tqdm(df_east.iterrows()):
    t = eval(row['news_token_bert'])
    if t['phrases'][0]['sentiment'] == '利好':
        df_east.loc[i,('sentiment')] = 1
    elif t['phrases'][0]['sentiment'] == '利空':
        df_east.loc[i,('sentiment')] = -1
    else:
        df_east.loc[i,('sentiment')] = 0
```

```python
def write_title(name):
    file_name = "./data/company/title_" + name + ".txt"
    title_name = name+"_title"
    with open(file_name, "w") as f:
        for title in eval(title_name):
            f.write(title+"\n")
```

```python
write_title("tesla")
write_title("microsoft")
write_title("jd")
write_title("vanke")
write_title("zzd")
```
